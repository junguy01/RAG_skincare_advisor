# Save a skincare routine guide from instagram
* `xskincare` is one of the biggest German skincare influencers and I found this post from him with a skincare guide
* we save the information in .jpg format

```python
from pathlib import Path
import re
import sys
import instaloader
```

```python
POST_URL = "https://www.instagram.com/p/CvUO2zWK_N-/?hl=de&img_index=1"
OUTPUT_DIR = Path("instagram_downloads")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
```

```python
def extract_shortcode(url: str) -> str:
    match = re.search(r"/p/([^/]+)/", url)
    if not match:
        raise ValueError("Konnte keinen Instagram-Shortcode aus der URL lesen.")
    return match.group(1)
```

```python
shortcode = extract_shortcode(POST_URL)
shortcode
```

```python
loader = instaloader.Instaloader(
    dirname_pattern=str(OUTPUT_DIR / "{target}"),
    filename_pattern="{date_utc}_UTC",
    download_video_thumbnails=False,
    save_metadata=True,
    post_metadata_txt_pattern="{caption}",
)
```

```python
post = instaloader.Post.from_shortcode(loader.context, shortcode)
target_folder = f"{post.owner_username}_{shortcode}"

loader.download_post(post, target=target_folder)

print("Gespeichert in:", OUTPUT_DIR / target_folder)
```
